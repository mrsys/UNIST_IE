##Seaborn 통해서 치료비 별 질병 진료 빈도를 알아보자
* 1000만원대 치료비가 드는 질병들이 있어서 그래프를 한번에 표현하기가 난해하여 나눠서 표현

```python
df_d = df_long['ED_RC_TOT_AMT'].sort_values()
print(df_d)
counts = df_d.value_counts()
counts = counts.sort_index()
print(counts)

#counts = df_general['ED_RC_TOT_AMT'].sort_values()

# x값이 가격
x_value = counts.index
x_value = x_value.tolist()
# y 값이 빈도
y_value = counts.values 
y_value = y_value.tolist()

x_0 = x_value[:500]
x_1 = x_value[500:1000]
x_2 = x_value[1000:2500]
x_3 = x_value[2500:5000]
x_4 = x_value[5000:7500]
x_5 = x_value[7500:9000]
x_6 = x_value[9000:10000]
x_7 = x_value[10000:11500]
x_8 = x_value[11500:]

y_0 = y_value[:500]
y_1 = y_value[500:1000]
y_2 = y_value[1000:2500]
y_3 = y_value[2500:5000]
y_4 = y_value[5000:7500]
y_5 = y_value[7500:9000]
y_6 = y_value[9000:10000]
y_7 = y_value[10000:11500]
y_8 = y_value[11500:]
```

* 아래 그래프는 초반 6만원 이하의 질병

```python
import seaborn as sns
sns.set_theme(style="darkgrid")
sns.set(rc = {'figure.figsize':(15,8)})
# Plot the responses for different events and regions
sns.lineplot(x=x_0, y=y_0,         
            )
# Plot the responses for different events and regions
sns.lineplot(x=x_1, y=y_1,             
            )
# Plot the responses for different events and regions
sns.lineplot(x=x_2, y=y_2,             
            )
sns.lineplot(x=x_3, y=y_3,         
            )
```
