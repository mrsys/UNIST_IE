```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
import pandas as pd
import numpy as np
import unicodedata
import os
import time
from tqdm import tqdm
from sklearn.utils import shuffle
import seaborn as sns
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') 
from matplotlib import rcParams

data_dir = '/content/drive/My Drive/Colab Notebooks/프로젝트랩/data/월별데이터'
```

```python
import glob
all_files = glob.glob(data_dir + "/*.csv")
all_files

index1 = all_files.index(all_files[0]); index2 = all_files.index(all_files[1])
all_files[index1], all_files[index2] = all_files[index2], all_files[index1]
all_files
```

```python
all_data = []
for filename in all_files:
  
    df = pd.read_csv(filename, index_col=None, header=0)
    """ 날짜 정보 넣기 """
    year = '20' + filename[-9:-7]
    month = filename[-6:-4]
    df['year'] = year
    df['month'] = month

    all_data.append(df)

frame = pd.concat(all_data, axis=0, ignore_index=True)
frame
```

```python
frame.to_csv(data_dir + 'clustering_data_1901_2004.csv')
```
